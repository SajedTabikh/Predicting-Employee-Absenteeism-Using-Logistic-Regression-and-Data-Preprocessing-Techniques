```python
df = raw_csv_data.copy()
df
```

### Drop the ID Column

```python
# Drop the id column as it's not relevant for the analysis:
df = df.drop('ID', axis=1)
df
```

```python
#list all the reason for absence (unique and in order ):
reason_for_absence_list = df['Reason for Absence'].unique()
reason_for_absence_list.sort()
reason_for_absence_list
# So e can see that we have 28 unique reasons for absence. But we have 1 missing reason for absence.(nb: 20 is not exist in the data)
```

### Make dummies for the Reason of absence column

```python
# Make dummies for the Reason of absence column:
# We use drop_first=True to drop the first column of the dummies dataframe, to avoid multicollinearity
reasons_columns = pd.get_dummies(df['Reason for Absence'], drop_first= True,dtype=int)
reasons_columns
```

```python
# Add a colum called check to see if the absence is just for one reason or for multiple reasons:
reasons_columns['Check'] = reasons_columns.sum(axis=1)
reasons_columns
```

```python
# We add the check column to check wether the absence is for one reason or for multiple reasons so after we conclude that it's not a missing value, 
# and each row have only one reason so now we can remove this column from the table:
reasons_columns = reasons_columns.drop('Check', axis=1)
reasons_columns
```

### Group the reasons for absence

```python
df = df.drop(['Reason for Absence'], axis = 1)
df
```

```python
# df.columns.values to get all the column names:
df.columns.values
```

```python
reasons_columns.columns.values
```

```python
# Split the reasons for absence into multiple dummy variables, and then group them in the following way: 
# Group 1: Columns 1 to 14 , 
# Group 2: Columns 15, 16, and 17, 
# Group 3: Columns 18, 19, 20, and 21,
# Group 4: Columns 22 to 28

reason_type_1 = reasons_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reasons_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reasons_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reasons_columns.loc[:, 22:28].max(axis=1)


# reason_type_0: No reason for absence
# reason_type_1: Various Diseases
# reason_type_2: Pregnancy and giving birth
# reason_type_3: Poisoning
# reason_type_4: Light Diseases
```

```python
reason_type_2
```

```python
# Concatenate the dummies with the original dataframe:
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
```

### Change the column names to meaningful names for the last added columns (0,1,2,3)

```python
df.columns.values
```

```python
columns_names = ['Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', "Reason_1", "Reason_2", "Reason_3", "Reason_4"]

df.columns = columns_names
df.head()
```

### Reorder Columns

```python
# Reorder the columns in the dataframe:

column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                         'Date', 'Transportation Expense', 'Distance to Work', 'Age',
                         'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
                         'Pets', 'Absenteeism Time in Hours'
                         ]

df = df[column_names_reordered]
df.head()
```

### Create a checkpoint

```python
# Create a checkpoint to save the processed dataframe:
df_reason_mod = df.copy()
df_reason_mod
```

### Date preprocessing

```python
df_reason_mod['Date']
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
df_reason_mod['Date']
```

```python
df_reason_mod.info()
# We successfully converted the 'Date' column to datetime format.
```

### Extract the Month value

```python
# Extract the Month values from the entire 'Date' columns and store it in a list called list_months and then add it as a new column in the dataframe called Month Value:

list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)

df_reason_mod['Month Value'] = list_months

df_reason_mod.head(20)
```

### Extract the Day of the Week value

```python
# Extract the Day of the Week value 
def date_to_weekday(date_value):
    return date_value.weekday()

df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head(20)
# 0 means monday, 1 means tuesday,..., 6 means sunday.
```

```python
df_reason_mod.columns.values
```

```python
# change the order of the columns:

column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                         'Date', 'Month Value', 'Day of the Week',
                         'Transportation Expense', 'Distance to Work', 'Age',
                         'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
                         'Pets', 'Absenteeism Time in Hours'
                         ]

df_reason_mod = df_reason_mod[column_names_reordered]
```

```python
# Remove the date from the dataframe:
df_preprocessed = df_reason_mod.drop('Date', axis=1)
```

### We wont manipulate the next 5 columns (Transportation expenses -> Body Mass index) as they are already in a good format.

```python
# Extract the unique values of the education column:
df_reason_mod['Education'].unique()
# 1: High School, 2: Graduate, 3: Postgraduate, 4: Master or Doctor
```

```python
# Count how many 1s, 2s, 3s, and 4s are in the Education column:
df_reason_mod['Education'].value_counts()
```

```python
# We have almost 600 observation that have only high school diploma and the rest have either graduate, postgraduate, or master or doctor degrees.
# So its not relevant to have 4 categories. Instead we can combine high school diploma in a group and graduate, postgraduate, and master or doctor in other group.
# So we should map the data and set the value of 1 is 0 and the value of 2, 3, 4 is 1:

df_reason_mod['Education'] = df_reason_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```
